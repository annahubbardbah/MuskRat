```python
#Function ingests a Lists, returns the most frequent value in list
def most_frequent(List):
    counter = 0
    num = List[0]
     
    for i in List:
        curr_frequency = List.count(i)
        if(curr_frequency> counter):
            counter = curr_frequency
            num = i
 
    return num
```

```python
def display_img_with_data():    
    for idx, img in enumerate(model_data):
        #print(len(model_data))
        #print(idx)
        img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
        #show image
        plt.imshow(img, cmap="gray")
        plt.show()

        print("True Classification: ")
        print(CATEGORIES[true_class[idx]] + "\n")

        print("Aggregate Classification: ")
        print(CATEGORIES[most_freq[idx]]+ "\n")

        print("Individual Model(s) Classification: ")
        #Evaluate all results of all 5 models 
        for j in range(len(total_pred[idx])):
                print("Model " + str(j+1) + ": " + CATEGORIES[total_pred[idx][j]])
```

```python
#most_freq is the most frequent classification per image of all 5 models
#total pred is a list with all 5 model predictions in a sublist
#true_class is the true classification of the image
#ACCURACY = total correct num of predictions / total predictions
def accuracy_results():
    num_correct_sep = 0
    tot_sep = 0

    num_correct_all = 0
    tot_all = 0

    #Evaluate all results of all 5 models 
    for i in range(len(total_pred)):
        #print(total_pred[i])
        for j in range(len(total_pred[i])):
            #increment num of prediction
            tot_sep = tot_sep + 1
            #print(total_pred[i][j])
            if(total_pred[i][j] == true_class[i]):
                #increment num correct prediction, if it equals true_class
                num_correct_sep = num_correct_sep + 1

    #evaluate most common result of the 5 models
    for i in range(len(most_freq)):
        tot_all = tot_all + 1
        if(most_freq[i] == true_class[i]):
            num_correct_all = num_correct_all + 1

    print("Accuracy when the Classification is the Result of One Model")
    #print(num_correct_sep)
    #print(tot_sep)
    temp = (num_correct_sep/tot_sep)*100
    print(str(temp) + " %")

    print("Accuracy when the Classification is the Most Frequent Result of 5 Models")
    #print(num_correct_all)
    #print(tot_all)
    temp = (num_correct_all/tot_all)*100
    print(str(temp) + " %")
```

```python
display_img_with_data()
accuracy_results()
```
